```python
import pandas as pd
import statsmodels.api as sm
import seaborn as sns
from scipy import stats as sts
import matplotlib.pyplot as plt
from statsmodels.stats.stattools import durbin_watson as dwtest
from statsmodels.stats.multicomp import pairwise_tukeyhsd
import numpy as np
!pip install pingouin --quiet
import pingouin as pg
```

##Data

```python
path = "/content/drive/MyDrive/Data/data_q.xlsx" # from drive
# path = "Data/data_q.xlsx" # from the computer
```

##Caffeine Anova & tukey

```python
df_cf= pd.read_excel(path) # data frame to be used for caffeine analysis
df_cf
```

```python
formula = 'caffeine	 ~  location * light * age'
model = sm.formula.ols(formula, data= df_cf).fit()
```

```python
print(model.summary())#model information
```

#Normality Tests

##Shapiroâ€“Wilk test

```python
w, p= sts.shapiro(model.resid)
print('w=%.3f, p_value=%.3f' % (w, p))
```

#Levene test for homogeneity of variance

```python
df_cf['interaction'] = df_cf['location']+ df_cf['light']+ df_cf['age']

pg.homoscedasticity(data= df_cf,
                    dv='caffeine',
                    group='interaction').round(5)
```

#Assumption of independence:

# Durbin-Watson Test

```python
dwtest(resids=np.array(model.resid))
```

#Anova

```python
aov_table =  sm.stats.anova_lm(model,typ=1)
aov_table
```
